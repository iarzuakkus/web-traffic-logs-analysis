```python
df = df.dropna(subset=['Timestamp'])
```

```python
df = df.dropna(subset=['HTTP_Method'])
```

Response_Size sutununundaki NaN degerleri ortalamasi ile doldurduk

```python
df['Response_Size'] = df['Response_Size'].fillna(df['Response_Size'].mean())
```

```python
df.info() # Veri setimizde artik NaN degerler bulunmamakta
```

```python
# Zaman damgalarını datetime formatına dönüştürme
df['Timestamp'] = pd.to_datetime(df['Timestamp'], format='%d/%b/%Y:%H:%M:%S %z')
```
